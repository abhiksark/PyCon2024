```python
%load_ext Cython
```

```python
def addv0(x):
    y = 0
    for i in range(x):
        for j in range(x):
            y += i
    return y
```

```python
%%cython -a
def addv1(x):
    y = 0
    for i in range(x):
        for j in range(x):
            y += i
    return y
```

```python
import timeit

##use timeit.timeit to compare the time taken by the two functions
v0 = timeit.timeit('addv0(100)',setup='from __main__ import addv0',number=1000)
v1 = timeit.timeit('addv1(100)',setup='from __main__ import addv1',number=1000)

print(v0, v1)   
print('Cython is {}x faster'.format(v0/v1))
```

```python
%%cython -a
def addv2(int x):
    cdef int i
    cdef int j
    cdef int y=0
    for i in range(x):
        for j in range(x):
            y += i
    return y
```

```python
v2 = timeit.timeit('addv2(100)',setup='from __main__ import addv2',number=1000)
print(v2)
print('Cython is {}x faster'.format(v0/v2))
```

```python
%%cython -a
cpdef int addv3(int x):
    cdef:
        int i, j, y = 0

    for i in range(x):
        for j in range(x):
            y += i
    return y
```

```python
v4 = timeit.timeit('addv3(100)',setup='from __main__ import addv3',number=1000)
print(v4)
print('Cython is {}x faster'.format(v0/v4))
```

```python
%%cython -a
cdef int addv4_cy(int x):
    cdef:
        int i, j, y = 0
        
    for i in range(x):
        for j in range(x):
            y += i
    return y

def addv4(x):
    return addv4_cy(x)
```
